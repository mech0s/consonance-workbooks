```python
from pytuning.visualizations import consonance_matrix
```

```python
edo_scale_12 = create_edo_scale(12)
```

```python
def metric_fifth(degree):
    p5 = sp.Rational(3,2)
    normalized_degree = normalize_interval(degree)
    y = normalize_interval(p5 / normalized_degree).evalf()
    y = y if y == 1 else 0
    return y
```

```python
def metric_test(degree):
    if (degree < 1):
        return 0
    else:
        return degree * 100
    
```

```python
matrix12 = consonance_matrix(edo_scale_12, metric_function=metric_test , figsize = (20,20), )
```
