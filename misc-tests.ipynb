```python
edo_scale_12 = create_edo_scale(12)
```

```python
def metric_test(degree):
    return normalize_interval(degree) * 100
    #return degree * 100
```

```python
def metric_fifth(degree):
    p5 = sp.Rational(3,2)
    normalized_degree = normalize_interval(degree)
    y = normalize_interval(p5 / normalized_degree).evalf()
    y = y if y == 1 else 0
    return y
```

```python
matrix12 = consonance_matrix(edo_scale_12, metric_function=metric_test , figsize = (20,20))
```

```python
scale = pytuning.scales.create_euler_fokker_scale([3,5],[2,1])
```

```python
matrixfokker = consonance_matrix(scale, metric_function=metric_fifth)
```
